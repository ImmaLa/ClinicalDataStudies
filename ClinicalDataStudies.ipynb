```python
df['Outcome Clean'] = df['Outcome Measures'].apply(clean_outcome)
```

```python
def clean_outcome(text):
    if pd.isna(text):
        return text
    
    text = text.strip().lower()

    if "overall survival" in text or text == "survival":
        return "Overall Survival"
    
    return text
```

```python
df = df.copy() ### Grouping trial status for a better visaulization
```

```python
df['Status Clean'] = df['Status'].replace(r'\s+', ' ', regex=True)
```

```python
Active = {
    "Recruiting",
    "Active, not recruiting",
    "Not yet recruiting",
    "Enrolling by invitation"
    "Available"
}

Completed = {
    "Completed"
}

Other = {
    "Terminated",
    "Suspended",
    "Withdrawn",
    "Approved for marketing",
    "Temporarily not available",
    "No longer available"
}
Unkown = {
    "Unknown status"
}
```

```python
def group_status(status):
    if status in Active:
        return "Active"
    elif status in Completed:
        return "Completed"
    elif status in Other:
        return "Other"
    else:
        return "Unknown"

df['Status Grouped'] = df['Status Clean'].apply(group_status)
```

```python
df['Status Grouped'].value_counts()
```

```python
status_pct = df['Status Grouped'].value_counts(normalize=True) * 100

plt.figure(figsize=(8,5))

plt.bar(
    status_pct.index,
    status_pct.values,
    color='orange'
)

plt.title("Trial Status Distribution (%)")
plt.xlabel("Status Grouped")
plt.ylabel("Percentage")

# Add labels on bars
for i, v in enumerate(status_pct.values):
    plt.text(i, v + 0.5, f"{v:.1f}%", ha='center')

plt.tight_layout()
plt.show()
```

#### The distribution shows that Completed trials form the largest group, indicating that most studies have reached completion and outcome reporting stages, followed by Active trials which represent a strong ongoing research pipeline.

#### Meanwhile, Other and Unknown categories highlight a smaller but important portion of terminated, suspended, or unclassified studies, pointing to areas where data completeness and trial status tracking could be improved.

```python
import nbformat ####Extratings the codes

def extract_code_from_ipynb(ipynb_file, output_file):
    with open(ipynb_file, 'r', encoding='utf-8') as file:
        notebook = nbformat.read(file, as_version=4)
        
    code_cells = [cell['source'] for cell in notebook['cells'] if cell['cell_type'] == 'code']
    
    with open(output_file, 'w', encoding='utf-8') as file:
        for i, code in enumerate(code_cells, 1):
            file.write(code)
            file.write('\n\n')
```

```python
extract_code_from_ipynb('ClinicalDataStudies.ipynb', 'ClinicalDataStudies.py')
```
